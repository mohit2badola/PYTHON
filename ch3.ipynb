```python
class car:
    def val(self,tyre,weight,engine):
        self.tyre= tyre
        self.weight=weight
        self.engine=engine
    def display(self):

        mileage = 28 - 4 * self.engine - 0.006 * (self.weight - 1000) - 0.7 * (self.tyre - 15)
        print(mileage  )
    
c1=car()
c1.val(15,1200,1.2)
c1.display()
```
