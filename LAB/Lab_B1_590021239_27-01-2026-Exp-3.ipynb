### ***1. Check whether a given number is divisible by 3 and 5 both.***

```python
num=int(input("Enter the number to check the divisibility by 3 and 5"))

if num%3==0:
    if num%5==0:
        print("The number is divisible by 3 and 5 both")
    else:
        print("The number is divisible by 3 only")
else:
    print("The number is divisible by none of them")

# Input given 15
```

```python
num=int(input("Enter a number to check the divisibility by 5"))
if num%5==0:
    print(f"The number is a multiple of 5 with the value multiple of {num/5}")
else:
    print("The number is not a multiple of 5")
# Input 14565
```

### ***3. Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.***

```python
num1=int(input("Enter the num 1"))
num2=int(input("Enter the num 2"))

if num1>num2:
    print("Number 1 is greater than number 2")
elif num2>num1:
    print("Number 2 is greater than number 1")
else:
    print("Both the numbers are equal")

##Input 45,32
```

### ***5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots***

```python
a=int(input("Enter the value of a "))
b=int(input("Enter the value of b "))
c=int(input("Enter the value of c "))

discrt=(b**2)-(4*a*c)

if discrt>0:
    print("Real Roots")
    
    root1=(-b+(discrt**0.5))/(2*a)
    root2=(-b-(discrt**0.5))/(2*a)

    print(f"Root  1 : {root1}")
    print(f"Root  2 : {root2}")
elif discrt<0:
    print("The root are imaginary")
else:
    print("Both root are same and real")
```

### ***6. Find whether a given year is a leap year or not.***

```python
year=int(input("Enter the year"))

if (year %4)==0 and year%400==0:
    print("the year is a leap year") 
else:
    print("Not a leap year")
```
