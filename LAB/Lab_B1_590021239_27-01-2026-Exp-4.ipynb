### ***1. Find a factorial of given number.***

```python
def factorial(num):
    if num==0 or num==1:
        return 1
    else:
        return num*factorial(num-1)

num=int(input("Enter the number "))

if num>=0:
    print(f"The factorial for {num} is {factorial(num)}")
else:
    print("Factorial of the negatice number is not possible")
```

### ***2. Find whether the given number is Armstrong number.***

```python
num =int(input("Enter a number: "))

original = num
num_digits = len(str(num))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** num_digits
    num //= 10

if sum == original:
    print(f"{original} is an Armstrong number")
else:
    print(f"{original} is not an Armstrong number")
```

### ***3. Print Fibonacci series up to given term.***

```python
num = int(input("Enter number of terms: "))

a = 0
b = 1

if num <= 0:
    print("Please enter a positive number")
elif num == 1:
    print(a)
else:
    print("Fibonacci Series:")
    print(a, b, end=" ")
    for i in range(2, num):
        c = a + b
        print(c, end=" ")
        a = b
        b = c
```

### ***4. Write a program to find if given number is prime number or not.***

```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")

```

### ***5. Check whether given number is palindrome or not.***

```python
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

### ***6. Write a program to print sum of digits.***

```python
num = int(input("Enter a number: "))

sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num //= 10

print("Sum of digits:", sum)
```

### ***7. Count and print all numbers divisible by 5 or 7 between 1 to 100.***

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nTotal count:", count)
```
