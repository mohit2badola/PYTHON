### 10. Write a program to print truth table for bitwise operators (&, |, and ^ operators)  


```python
print("--------------------")

print("|A B  | A&B A|B A^B|")
print("|-----|------------|")

for a in (0, 1):
    for b in (0, 1):
        print("|",a, b, "|", a & b, " ", a | b, " ", a ^ b, " |")

print("--------------------")

```
