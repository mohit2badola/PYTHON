### ***10. Write a program to print truth table for bitwise operators (&, |, and ^ operators)*** 


```python
print("--------------------")

print("|A B  | A&B A|B A^B|")
print("|-----|------------|")

for a in (0, 1):
    for b in (0, 1):
        print("|",a, b, "|", a & b, " ", a | b, " ", a ^ b, " |")

print("--------------------")

```

### ***11. Write a program to find left shift and right shift values of a given number.***  


```python
num=int(input("Enter a number: "))
s=int(input("Enter number of positions to shift: "))
ls = num << s
rs = num >> s
print("Left Shift Value:", ls)
print("Right Shift Value:", rs)
```
