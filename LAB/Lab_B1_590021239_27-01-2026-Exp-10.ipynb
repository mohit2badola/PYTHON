1.	Create numpy array to find sum of all elements in an array. 

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

total = np.sum(arr)

print("Array:", arr)
print("Sum of all elements:", total)
```

2.	Create numpy array of (3,3) dimension. Now find sum of all rows & columns individually. Also find 2nd maximum element in the array.  

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

print("Array:\n", arr)

# Row sum
print("Row Sum:", np.sum(arr, axis=1))

# Column sum
print("Column Sum:", np.sum(arr, axis=0))

# 2nd Maximum
flat = arr.flatten()
second_max = np.sort(flat)[-2]

print("Second Maximum Element:", second_max)
```

3.	Perform Matrix multiplication of any 2 n*n matrices. 

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```

4.	Write a Pandas program to get the powers of an array values element-wise.  
Note: First array elements raised to powers from second array 
Sample data: {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86],'Z':[86,97,96,72,83]} 
Expected Output: 
X Y Z 
0 78 84 86 
1 85 94 97 
2 96 89 96 
3 80 83 72 
4 86 86 83 


```python
import pandas as pd

data = {'X':[78,85,96,80,86],
        'Y':[84,94,89,83,86],
        'Z':[86,97,96,72,83]}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

# Raise each element to power 2
power_df = df.pow(2)

print("Element-wise Power:\n", power_df)
```

6.	Write a Pandas program to find and replace the missing values in a given DataFrame which do not have any valuable information. 

```python
import pandas as pd
import numpy as np

data = {'A':[1,2,np.nan,4],
        'B':[5,np.nan,7,8]}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)

# Replace missing values with 0
df_filled = df.fillna(0)

print("After Replacing Missing Values:\n", df_filled)
```


7.	Create a program to demonstrate different visual forms using Matplotlib.


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,25,30,40]

# Line Chart
plt.plot(x, y)
plt.title("Line Chart")
plt.show()

# Bar Chart
plt.bar(x, y)
plt.title("Bar Chart")
plt.show()

# Pie Chart
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()

# Scatter Plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```
